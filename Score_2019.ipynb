```python
import numpy as np
import random

def eudist(x,y): #ระยะทาง Euclidean
  d = np.sqrt((x-y)**2)
  return d

x = df
k = 2

k1 = random.randint(1,100)
k2 = random.randint(1,100)

print(k1)
print(k2)

p = 0
while(1):
  p+=1
  l1 = [ ]
  l2 = [ ]
  for v in x:
    dist1 = eudist(k1,v)
    dist2 = eudist(k2,v)
    if(dist1 <= dist2):#ถ้า dist1 น้อยกว่า dist2
      l1.append(v)#เพิ่มข้อมูลลงในตัวแปร List l1
    else:
      l2.append(v)#เพิ่มข้อมูลลงในตัวแปร List l2
  m1 = np.mean(l1)#หาค่าเฉลี่ย l1
  m2 = np.mean(l2)#หาค่าเฉลี่ย l2

  if(m1 == k1 and m2 == k2): #การหยุด
    break
  else:
    k1 = m1
    k2 = m2
    print("round: "+str(p))
    print(l1)
    print(l2)
    print("-----------------------------------")
```
