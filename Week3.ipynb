```python
import numpy as np
from random import randint

def eudist(x,y): #ระยะทาง Euclidean
  d = np.sqrt((x-y)**2)
  return d

x = [1,7,15,23,17,15,22,74,25,12,16,44,75,62,34] #กำหนดตัวเลข
k = 2

k1 = 32 #ที่จริง k1 คือ เลขจะเกิดจากการสุม
k2 = 12 #ที่จริง k2 คือ เลขจะเกิดจากการสุม

p = 0
while(1):
  p+=1
  l1 = [ ]
  l2 = [ ]
  for v in x:
    dist1 = eudist(k1,v)
    dist2 = eudist(k2,v)
    if(dist1 <= dist2):#ถ้า dist1 น้อยกว่า dist2
      l1.append(v) #เพิ่มข้อมูลลงในตัวแปร List l1
    else:
      l2.append(v)#เพิ่มข้อมูลลงในตัวแปร List l2
  m1 = np.mean(l1)#หาค่าเฉลี่ย l1
  m2 = np.mean(l2)#หาค่าเฉลี่ย l2

  if(m1 == k1 and m2 == k2): #การหยุด
    break
  else:
    k1 = m1
    k2 = m2
    print("round: "+str(p))
    print(l1)
    print(l2)
    print("-----------------------------------")
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt

def dist(x,y):
  d = np.sqrt(np.sum((np.array(x) - np.array(y))**2))
  return d

#print(dist([1,1], [5,5]))

data = make_blobs(n_samples=100, n_features=2, centers =4, cluster_std=3, random_state=101) #กำหนด data

#print(data[0][0])
#print(data[0][1])

k = 2
k1 = [15,13 ]
k2 = [0, 22 ]

p = 0
e = 0.02

while(1):
  p+=1
  l1 = [ ]
  l2 = [ ]
  for v in data[0]:
    d1 = dist(k1,v)
    d2 = dist(k2,v)
    if(d1 <= d2):
      l1.append(v)
    else:
      l2.append(v)
  m1 = np.mean(np.array(l1), axis=0)
  m2 = np.mean(np.array(l2), axis=0)
  if dist(m1,k1) < e and dist(m2,k2) < e :
    break
  else:
    k1 = m1
    k2 = m2
    print("round: "+str(p))
    print(l1)
    print(l2)
    print("-----------------------------------")
    plt.scatter(np.array(l1)[:,0],np.array(l1)[:,1], c='red')
    plt.scatter(np.array(l2)[:,0],np.array(l2)[:,1], c='blue')
    plt.show()

print(p)
```

```python
import cv2 
from google.colab.patches import cv2_imshow
from sklearn.cluster import KMeans

img = cv2.imread('/content/drive/MyDrive/DS/iris-hotel.jpg',1)#อ่านภาพ
#cv2_imshow(img)
#print(img[0][0])
Kmean = KMeans(n_clusters = 4)
row, col , d = img.shape

dat = img.reshape(1, row*col,d)
output = Kmean.fit(dat[0])
cmap = [[120,15,200], [255,255,255], [25,25,25],[1,120,255]]

dis = np.zeros((row,col,3), dtype = np.uint8)
k = 0
for i in range (row):
  for j in range(col):
    dis[i,j] = cmap[int(output.labels_[k])]
    k+=1

cv2_imshow(dis)
```
