```python
from keras.datasets import fashion_mnist
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()

train_x = np.zeros((x_train.shape[0], 28 * 28))
for i,x in enumerate(x_train):
  train_x[i] = x.reshape(1,28 * 28)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(train_x, y_train)
```

```python
test_x = np.zeros((x_test.shape[0], 28*28))
for i,x in enumerate(x_test):
  test_x[i] = x.reshape(1,28 * 28)
predict = knn.predict(test_x[0 : 50])
true = y_test[0 : 50]
print(predict)
print(true)
```

```python
cm = np.zeros((10,10))
for i in range(50):
  p = predict[i]
  t = true[i]
  cm[p][t] = cm[p][t] + 1
print(cm)
print(sum(sum(cm)))
total = sum(sum(cm))

acc_sum = 0
acc = 0
for i in range(10):
  acc_sum += cm[i][i]
acc = acc_sum/total
print('accuaracy : ' + str(acc))

precision = np.zeros((10,1), dtype = np.float32)
recall = np.zeros((10,1), dtype = np.float32)
total_p = 0
total_c = 0

for i in range(10):
  precision[i] = cm[i][i]/sum(cm[i,:])
  recall[i] = cm[i][i]/sum(cm[:,i])
  total_p += 1/precision[i]
  total_c += 1/precision[i]

l_precision = 10/total_p
l_recall= 10/total_c
#print(l_precision, l_recall)

f1 = 2 * (l_precision * l_recall) / (l_recall + l_precision)
#print(f1)

print('precision : ' + str(l_precision[0]))
print('recall : ' + str(l_recall[0]))
print('f1 : ' + str(f1[0]))

#print(precision)
#print('------------------------------')
#print(recall)
```
